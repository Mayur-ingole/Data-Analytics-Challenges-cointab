```python
def weightslab(x):
    if 0 < x <= 0.5:
        return  0.5
    elif 0.5 < x <= 1:
        return 1
    elif 1 < x <= 1.5:
        return 1.5
    elif 1.5 < x <= 2:
        return 2
    elif 2 < x <= 2.5 :
        return 2.5
    elif 2.5 < x <= 3:
        return 3
    elif 3 < x <= 3.5 :
        return  3.5
    elif 3.5 < x <= 4:
        return  4
    elif 4 < x <= 4.5:
        return  4.5
    elif 4.5 < x <= 5 :
        return  5
```

```python
df3["Weight slab as per X (KG)"] = df3["Total weight as per X (KG)"].apply(weightslab)
df3["Weight slab charged by Courier Company (KG)"] = df3["Total weight as per X (KG)"].apply(weightslab)
```
